```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
titanic_data = sns.load_dataset('titanic')
```

```python
print(titanic_data.head(891))
```

```python
print(titanic_data)
```

```python
numeric_columns = titanic_data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = titanic_data[numeric_columns].corr()
```

```python
group = titanic_data.groupby(['class','survived'],observed=False)

pclass_survived = group.size().unstack()

sns.heatmap(pclass_survived,cmap='coolwarm',annot = True, fmt ="d")
```

```python
titanic_data['survived'] = titanic_data['survived'].replace({0: 'Survived', 1: 'Not Survived'})

sns.catplot(x='sex',hue='survived', palette='dark:red',kind='count',data=titanic_data)
```

```python
sns.violinplot(x ="sex", y ="age", hue ="survived",palette='dark:red', data = titanic_data, split = True)
```

```python
titanic_data['Fare_Range'] = pd.qcut(titanic_data['fare'], 4)
 
# Barplot - Shows approximate values based 
# on the height of bars.
sns.barplot(x ='Fare_Range', y ='survived', data = titanic_data)
```

```python
sns.catplot(x ='embarked', hue ='survived', kind ='count', col ='class', data = titanic_data)
```

```python
sns.histplot(data=titanic_data, x='fare')
```

```python
sns.boxplot(x=titanic_data['sex'], y=titanic_data['age'],palette='dark:red' ,hue=titanic_data['survived'])
```
