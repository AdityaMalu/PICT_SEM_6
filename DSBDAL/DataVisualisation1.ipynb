```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from seaborn import set_style, countplot, lineplot, scatterplot, boxplot, pointplot, violinplot, swarmplot, barplot, kdeplot 
```

```python
data = sns.load_dataset('titanic')
set_style("whitegrid")
```

```python
data.dtypes
```

```python
data['survived'] = data['survived'].replace({0:'Survived' , 1:'Not Survived'})
sns.catplot(data = data,hue = 'survived', x = 'sex' , kind='count')
```

```python
# Histogram
group = data.groupby(['class','survived'], observed=False)
pclass_survived = group.size().unstack()
sns.heatmap(pclass_survived,annot=True,fmt='d')
```

```python
# Line plot
sns.violinplot(data=data,x='age',y='sex',hue = 'survived',split=True)
```

```python
# Scatter plot
data['fare_range'] = pd.qcut(data['fare'],4)
sns.barplot(data=data,x = 'fare_range' , y = 'survived')
```

```python
# Box plot
sns.catplot(data = data, x = 'embarked' , hue='survived' , kind='count' , col='class')
```

```python
# Point plot
sns.histplot(data=data, x = 'fare', kde=True)
```

```python
# Count plot
sns.boxplot(data=data, x='age' , y = 'sex' , hue='survived')
```

```python
# Violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(data=data, x='fare')
plt.title('Violin Plot of Fare')
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(8, 6))
sns.barplot(data=data, x='fare', estimator=sum)
plt.title('Bar Plot of Fare')
plt.show()
```

```python
# KDE plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data=data['fare'], shade=True)
plt.title('KDE Plot of Fare')
plt.show()
```
